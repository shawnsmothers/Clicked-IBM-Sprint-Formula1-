```python
results = pd.read_csv('/content/results.csv')
```

```python
# replace position n\ with NaN
results['position'] = results['position'].replace('n\a', np.nan, regex = True)
```

```python
from sre_constants import error
# convert to numeric
results['position']= pd.to_numeric(results['position'], errors = 'coerce')
```

```python
# calculate the mean of the position
mean_position = results['position'].mean()
```

```python
# get rounded mean and fill NaN
results['position'] = results['position'].fillna(round(mean_position))
```

```python
# convert datatype into int
results['position'] = results['position'].astype(int)
```

```python
# caluclate the mean of points exluding 0
mean_points = results['points'].replace(0, np.nan).mean()
```

```python
# replace zeros with the mean
results['points'] = results['points'].replace(0, mean_points)
```

```python
results['points'] = results['points'].round(2)
```
